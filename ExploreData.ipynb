```python
import glob
traj_files = glob.glob('traj_data/talented-steel-butterfly*.csv')
```

```python
traj_files
```

```python
import pandas as pd
import numpy as np
res = 0.01
for traj_file in traj_files:
    # with open(traj_files[0], 'r') as f:
    row = pd.read_csv(traj_file)
    errors = []
    for i in range(200):
        dX = np.array([row.iloc[i+1]['dX'], row.iloc[i+1]['dY'], row.iloc[i+1]['dX1'], row.iloc[i+1]['dY1']])
        X1 = np.array([row.iloc[i]['X'], row.iloc[i]['Y'], row.iloc[i]['X1'], row.iloc[i]['Y1']])
        X2 = np.array([row.iloc[i+1]['X'], row.iloc[i+1]['Y'], row.iloc[i+1]['X1'], row.iloc[i+1]['Y1']])
        errors.append(list(X2-X1-dX*res))
    errors = np.array(errors)
    print(np.mean(errors, axis=0), np.std(errors, axis=0))
#     print(np.mean(errors), np.std(errors))
```
